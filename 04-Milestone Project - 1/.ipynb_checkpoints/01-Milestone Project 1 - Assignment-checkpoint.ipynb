```python
board = ['   *   *   ',[' ', '1', ' * ', '2', ' * ', '3'],'***********', [' ', '4', ' * ', '5', ' * ', '6'], '***********', [' ', '7', ' * ', '8', ' * ', '9'], '   *   *   ']
win = False
winner = ''
p1 = []
p2 = []
count = 0

def print_board():
    for line in board:
        if type(line) == list:
            temp = ''.join(line)
            print(temp)
        else:
            print(line)

def check_wins(playerInput):
    possibles = [[1,2,3],[4,5,6],[7,8,9],[1,4,7],[2,5,8],[3,6,9],[1,5,9],[3,5,7]]
    for thing in possibles:
        if thing[0] in playerInput and thing[1] in playerInput and thing[2] in playerInput:
            return True
    else:
        return False
def update_board(symbol, place):
    x = -1
    y = -1
    for line in board:
        x += 1
        if place in line:
            for item in line:
                y += 1
                if item == place:
                    board[x][y] = symbol

while win == False and count < 9:
    if count%2 == 0:
        p1Input = int(input('Player 1 move (please chose an avalible square by typing in the number):'))
        if p1Input not in p1 and p1Input not in p2:
            p1.append(p1Input)
            update_board('X', str(p1Input))
            win = check_wins(p1)
            print_board()
            if win == True:
                print('Player 1 is the winner!')
            count += 1
        else:
            print('Please enter a number 1-9 that has not been used.')
    else:
        p2Input = int(input('Player 2 move (please chose an avalible square by typing in the number):'))
        if p2Input not in p1 and p2Input not in p2:
            p2.append(p2Input)
            update_board('O', str(p2Input))
            win = check_wins(p2)
            print_board()
            if win == True:
                print('Player 2 is the winner!')
            count += 1
        else:
            print('Please enter a number 1-9 that has not been used.')

if count >= 9:
    print('Draw!')
```
